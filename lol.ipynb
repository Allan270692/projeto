```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("dados_aids_hiv_excel_1.xlsx")
df = df.dropna(subset=['NU_IDADE_N']) 

plt.hist(df['NU_IDADE_N'], color='pink', edgecolor='black') #criamos um histograma e personalizamos
intervalo = list(range(10, 120, 10)) #intervalo
plt.title("Ocorrência por idade") #atribui um título ao gráfico
plt.xlabel("Idade") #nomeia o eixo x
plt.xticks(intervalo) #eixo x terá os elementos descritos em intervalo
plt.ylabel("N° de Pessoas") #nomeia o eixo y

plt.savefig("distribuicao_idades.pdf", format='pdf') #gera um pdf da análise
plt.show() #exibe o gráfico
```
